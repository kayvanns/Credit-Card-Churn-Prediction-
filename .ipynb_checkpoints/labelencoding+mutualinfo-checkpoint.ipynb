```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.feature_selection import mutual_info_classif
```

```python
df = pd.read_csv("train.csv")
y = df['loan_defaulted']
```

```python
df_cleaned = df.copy()
categorical = df_cleaned.select_dtypes(include = "object")
categorical
```

```python
le=LabelEncoder()
encoded_gender = le.fit_transform(categorical["Gender"])
encoded_gender = pd.Series(encoded_gender)
encoded_marital = le.fit_transform(categorical["Marital_Status"])
encoded_marital = pd.Series(encoded_marital)
encoded_card = le.fit_transform(categorical["Card_Category"])
encoded_card = pd.Series(encoded_card)
encoded_card.value_counts()
categorical["Gender"] = encoded_gender
categorical["Marital_Status"] = encoded_marital
categorical["Card_Category"] = encoded_card
categorical
```

```python
education_order = ['Uneducated', 'High School', 'College', 'Graduate', 'Post-Graduate', 'Doctorate', 'Unknown']
income_order = income_order = ['Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +', 'Unknown']
oe = OrdinalEncoder(categories = [education_order,income_order])
ordinal = pd.DataFrame(oe.fit_transform(categorical[["Education_Level", "Income_Category"]]), columns = ["Education_Level","Income_Category"])
ordinal
```

```python
new_c = categorical.drop(columns = ["Education_Level", "Income_Category"])
```

```python
new_c
```

```python
ordinal
```

```python
final = pd.concat([ordinal, new_c], axis=1)
```

```python
final
```

```python
dff = df_cleaned.drop(columns = ["id","CLIENTNUM","loan_defaulted","Education_Level", "Income_Category", "Gender", "Marital_Status","Card_Category"])
```

```python
dff
```

```python
final_df = pd.concat([dff,final],axis=1)
```

```python
final_df
```

```python
final_df["Education_Level"] = final_df["Education_Level"].astype(int)
final_df["Income_Category"] = final_df["Income_Category"].astype(int)
```

```python
final_df
```

```python

mi_scores = mutual_info_classif(final_df, y, discrete_features='auto')
mi_series = pd.Series(mi_scores, index=final_df.columns).sort_values(ascending=False)
mi_series
```
