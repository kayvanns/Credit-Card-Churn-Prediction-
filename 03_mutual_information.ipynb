```python
import pandas as pd
from sklearn.feature_selection import mutual_info_classif
```

```python
df = pd.read_csv("train.csv")
X1= pd.read_csv("X1.csv")
X1 = X1.drop(columns="Unnamed: 0")
y = df.loan_defaulted
X1
```

```python

mi_scores = mutual_info_classif(X1, y, discrete_features='auto')
mi_series = pd.Series(mi_scores, index=X1.columns).sort_values(ascending=False)
mi_series
```
